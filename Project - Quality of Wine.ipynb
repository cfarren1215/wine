# Cleaning Red Wine Data

```python
data_red = pd.read_csv('winequality-red.csv')
data_red.head()
```

```python
# show all columns for red wine
data_red.columns
```

```python
# Adding a column for wine Type as Red for all type of red wine
data_red['Type'] = 'Red'

#extract 'sulphates', 'alcohol', 'quality' for red wine
reduced_red = data_red.loc[:, ['Type', 'alcohol', 'sulphates', 'alcohol', 'quality']]

# sort by quality from max to min

reduced_red.sort_values(by = ['quality'], ascending = False)
```

# Cleaning White Wine Data

```python
data_white = pd.read_csv('winequality-white.csv')
data_white.head()
```

```python
# show all columns for white wine
data_white.columns
```

```python
# Adding a column for wine Type as White for all types of white wine
data_white['Type'] = 'White'

#extract 'sulphates', 'alcohol', 'quality' for white wine
reduced_white = data_white.loc[:, ['Type', 'sulphates', 'alcohol', 'quality']]

# sort by quality from max to min

reduced_white.sort_values(by = ['quality'], ascending = False)
```

```python
combined_wine = pd.concat([reduced_white, reduced_red])
combined_wine.sort_values('quality', ascending=False, inplace=True, sort=False)
combined_wine.head()
```
